```python
import pandas as pd
import webbrowser
import time
#import pyautogui  # 用于模拟关闭浏览器标签页
#import keyboard  # 用于跨平台的键盘操作

# 读取Excel文件中的URL
excel_name = "小红书url.xlsx"
xls = pd.ExcelFile(excel_name)
sheet_names = xls.sheet_names
for sheet_name in sheet_names:
    print('处理sheet: {}'.format(sheet_name))
    df = pd.read_excel(excel_name, sheet_name=sheet_name, header=None, skiprows=1)
    urls = df[0].tolist()
    df['label'] = 1 # 1表明ok，0表明不ok
    for i in range(len(urls)):
        webbrowser.open(url=urls[i])

        # if (i + 1) % 10 == 0:
        #     pyautogui.hotkey('ctrl', 'w')

        # 判断label
        user_input = input("如果不好，输入0；若ok直接回车，若要退出，输入e：")
        if user_input.lower() == 'e':
            break  
        if user_input == '':
            df.loc[i, 'label'] = 1
        elif user_input == '0':
            df.loc[i, 'label'] = 0

    if user_input.lower() == 'e':
        break 
df.to_excel(f"{excel_name}_label.xlsx", index=False)
```

```python
# 时间戳转datetime
from datetime import datetime
def unix2datetime(timestamp_ms):
    # 毫秒为单位的时间戳
    timestamp_s = timestamp_ms / 1000.0  # 转换为秒
    dt_object = datetime.fromtimestamp(timestamp_s)

    print("时间戳:", timestamp_ms)
    print("转换后的datetime对象:", dt_object)
```
